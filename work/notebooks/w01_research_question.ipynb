## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

The code cell below loads the starter dataset and computes the share of declining pages and the median-impressions gap between declining and stable pages. As additional motivation already documented from this repo's own pipeline: on this 30,000-row starter slice, a plain rule-based baseline reached precision@50 of 0.240 (about 12 of the top 50 flagged pages were correct), while the trained random forest reached 0.740 (about 37 of 50) — evidence that a learned ranking meaningfully beats a fixed rule here, which is exactly the gap this lane exists to close.

**Actual computed numbers (run in Colab against the starter CSV):** the starter dataset has 30,000 rows and 44 columns; 54.2% of rows carry the `trend_direction == "down"` (declining) label; and declining pages have a notably higher median 90-day impressions (961) than non-declining pages (472). That last number is the interesting one for this lane: the pages most worth reviewing first aren't quiet, low-traffic pages — they are already-visible pages that are losing ground, which is exactly the kind of candidate a refresh-priority queue should surface.

```python
import pandas as pd
url = "https://raw.githubusercontent.com/Purvansh09/flyrannk_week1/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)
print("rows, cols:", df.shape)
print("declining share:", (df["trend_direction"] == "down").mean())
print("median impressions, declining vs not:")
print(df.groupby(df["trend_direction"] == "down")["impressions_90d"].median())
```
